```python
import pickle
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results_dir = './results/lifegate_0/'
```

```python
# Load the Q tables from the primary learning agent, Q_D and Q_R value functions
with open(results_dir+'tabular_ai.pkl', 'rb') as fq:
    ai = pickle.load(fq)
    
with open(results_dir+'tabular_qd.pkl', 'rb') as fd:
    ai_d = pickle.load(fd)
    
with open(results_dir+'tabular_qr.pkl', 'rb') as fr:
    ai_r = pickle.load(fr)
```

```python
q_table = np.zeros((10, 10, 5))
q_d = np.zeros_like(q_table)
q_r = np.zeros_like(q_table)
```

```python
for i in range(10):
    for j in range(10):
        for a in range(5):
            key = tuple([j, i, a])
            try:
                q_table[i,j,a] = ai.q[key]
                q_d[i,j,a] = ai_d.q[key]
                q_r[i,j,a] = ai_r.q[key]
            except:
                pass
```

```python
plt.imshow(np.max(q_table,axis=-1),vmin=-1., vmax=1., cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
plt.imshow(np.max(q_d, axis=-1),vmin=-1., vmax=1., cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
plt.imshow(np.max(q_r, axis=-1),vmin=-1., vmax=1., cmap='RdBu')
plt.colorbar()
plt.show()
```
